```python
# Akses data pdf tersebut pada direktori Google Colab
from tabula import read_pdf
df = read_pdf('Laporan_Nasional_RKD2018_FINAL.pdf', pages = 64)
```

```python
# Tampilkan data pdf dalam bentuk tabel (dataframe)
df[0]
```

```python
# Panggil library yang dibutuhkan
import numpy as np
# Hilangkan yang NaN dengan nilai kosong
df_edit = df[0].replace(np.nan, '')
df_edit
```

```python
# Gabungkan kolom pertama dengan kolom kedua
df_edit.iloc[:,0] = df_edit.iloc[:,0] + ' ' + df_edit.iloc[:,1]
df_edit
```

```python
# Hapus dua baris pertama dan baris paling akhir
df_edit = df_edit.iloc[2:-1].reset_index(drop=True)
df_edit
```

```python
# Hilangkan kolom kedua
df_edit = df_edit.iloc[:,[0,2,3,4,5,6,7]]
df_edit
```

```python
# Di bagian kanan data, tambahkan kolom yang diperlukan dan split kolom yang menyatu
df_edit['Mudah'] = df_edit.iloc[:,1]
df_edit[['Sulit', 'Sulit1', 'Sulit2']] = df_edit.iloc[:,3].str.split(' ',expand=True)
df_edit['Sangat_Sulit'] = df_edit.iloc[:,4]
df_edit
```

```python
# Ambil kolom yang diperlukan
df_edit = df_edit.iloc[:,[0,7,8,11]]
df_edit
```

```python
# Lakukan penamaan kolom pertama
df_edit = df_edit.rename(columns={
               df_edit.columns[0]:'Provinsi',
                        }
              )
df_edit
```

```python
# Sesuaikan nama provinsi agar dapat dijoin dengan data spasial (shapefile) 
# dengan merubah teks "JawaTimur" menjadi "Jawa Timur"
df_edit = df_edit.replace(to_replace='JawaTimur',value='Jawa Timur', regex=True)
df_edit
```

```python
# Export data ke csv, output akan terlihat pada direktori folder Google Colab (sebelah kiri), tunggu beberapa saat
df_edit.to_csv('Riskesdas_Python.csv')
```

**2. Join Table with Spatial Data**

```python
# Panggil library
import geopandas as gpd
```

```python
# Panggil data csv sebagai dataframe
sdf = gpd.read_file('Riskesdas_Python.csv')
sdf
```

```python
# Lakukan penamaan kolom pertama
sdf = sdf.rename(columns={
               sdf.columns[0]:'KODE',
                        }
              )
sdf
```

```python
# Ambil kolom yang diperlukan
sdf = sdf.iloc[:,[0,1,2,3,4]]
sdf
```

```python
# Ubah tipe data pada kolom-kolom data csv agar dapat dilakukan join dengan data spasial (shapefile)
sdf['KODE']=sdf['KODE'].astype(int)
sdf['Mudah']=sdf['Mudah'].astype(float)
sdf['Sulit']=sdf['Sulit'].astype(float)
sdf['Sangat_Sulit']=sdf['Sangat_Sulit'].astype(float)
```

```python
# Upload semua data 'Provinsi' dalam folder Provinsi (data spasial 'Provinsi.shp' dan ekstensi yang berbeda-beda)
# Klik "Choose Files" dan tunggu beberapa saat sampai semua data 'Provinsi' terupload
%cd /content/
from google.colab import files
files.upload()
```

```python
# Akses data spasial (shapefile) batas provinsi yang sudah diupload
spatial_data = gpd.read_file('Provinsi.shp')
spatial_data
```

```python
# Lakukan join table dari csv (dataframe) ke dalam data spasial dengan atribut "KODE" sebagai primary key
spatial_data = spatial_data.merge(sdf, on='KODE')
spatial_data
```

```python
# Ambil kolom yang dibutuhkan
spatial_data = spatial_data.iloc[:,[0,1,2,3,4,6,7,8]]
spatial_data
```

```python
# Perbaiki penamaan kolom Provinsi
spatial_data = spatial_data.rename(columns={
               spatial_data.columns[3]:'Provinsi',
                        }
              )
spatial_data
```

```python
# Export hasil join table ke shapefile (shp)
# Tunggu beberapa saat sampai hasil export 'Riskesdas_Provinsi' muncul di direktori sebelah kiri
spatial_data.to_file('Riskesdas_Provinsi.shp')
spatial_data
```
