```python
# Panggi library
import arcgis
import pandas as pd
```

```python
# Upload file csv (Tabula_Riskesdas_Python.csv) dari folder local computer, klik "Choose Files"
%cd /content/
from google.colab import files
files.upload()
```

```python
# Panggil data csv sebagai dataframe
dataset = 'Tabula_Riskesdas_Python.csv'
# panggil dataset sebagai dataframe
sdf = pd.read_csv(dataset)
sdf.head()
```

```python
# Ubah tipe data pada kolom-kolom data csv agar dapat dilakukan join dengan data spasial (shapefile)
sdf['Kode']=sdf['Kode'].astype(int)
sdf['Mudah']=sdf['Mudah'].astype(float)
sdf['Sulit']=sdf['Sulit'].astype(float)
sdf['Sangat_Sulit']=sdf['Sangat_Sulit'].astype(float)
```

```python
# Upload semua data 'Provinsi' dalam folder prov (data spasial 'Provinsi.shp' dan ekstensi yang berbeda-beda)
# Klik "Choose Files" dan tunggu beberapa saat sampai semua data 'Provinsi' terupload
%cd /content/
from google.colab import files
files.upload()
```

```python
# Panggil library
import geopandas as gpd
# Akses data spasial (shapefile) batas provinsi yang sudah diupload
spatial_data = gpd.read_file('Provinsi.shp')
spatial_data
```

```python
# Lakukan join table dari csv (dataframe) ke dalam data spasial dengan atribut "KODE" sebagai primary key
spatial_data = spatial_data.merge(sdf, left_on='KODE', right_on='Kode')
spatial_data
```

```python
spatial_data = spatial_data.iloc[:,[2,8,9,10,11,3,4,5,6]]
spatial_data
```

```python
# Export hasil join table ke shapefile (shp)
# Tunggu beberapa saat sampai hasil export 'Riskesdas_Provinsi' muncul di direktori sebelah kiri
spatial_data.to_file('Riskesdas_Provinsi.shp')
spatial_data
```
