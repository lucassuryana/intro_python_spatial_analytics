```python
from tabula import read_pdf
from tabulate import tabulate
```

```python
# Panggil datbel yang ada di dalam pdf di halaman 64
df = read_pdf('Laporan_Nasional_RKD2018_FINAL.pdf', pages = 64)
```

```python
df[0]
```

```python
import numpy as np
# Hilangkan yang NaN dengan nilai kosong
df_edit = df[0].replace(np.nan, '')
df_edit
```

```python
# Gabungkan kolom pertama dengan kolom kedua
df_edit.iloc[:,0] = df_edit.iloc[:,0] + ' ' + df_edit.iloc[:,1]
df_edit
```

```python
# Hapus dua baris pertama dan baris paling akhir
df_edit = df_edit.iloc[2:-1].reset_index(drop=True)
df_edit
```

```python
# Hilangkan kolom kedua
df_edit = df_edit.iloc[:,[0,2,3,4,5,6,7]]
df_edit
```

```python
df_edit[['1','2']] = df_edit.iloc[:,2].str.split(' ',expand=True)
df_edit[['3','4','5']] = df_edit.iloc[:,3].str.split(' ',expand=True)
df_edit[['7','8']] = df_edit.iloc[:,5].str.split(' ',expand=True)
df_edit
```

```python
df_edit = df_edit.iloc[:,[0,1,7,8,9,10,11,4,12,13,6]]
df_edit
```

```python
df_edit.columns[0]
```

```python
# rename columns
df_edit = df_edit.rename(columns={
               df_edit.columns[0]:'Provinsi',
               df_edit.columns[1]:'Mudah_%',
               df_edit.columns[2]:'Mudah_CI95%_1',
               df_edit.columns[3]:'Mudah_CI95%_2',
               df_edit.columns[4]:'Sulit_%',
               df_edit.columns[5]:'Sulit_CI95%_1',
               df_edit.columns[6]:'Sulit_CI95%_2',
               df_edit.columns[7]:'Sangat_Sulit_%',
               df_edit.columns[8]:'Sangat_Sulit_CI95%_1',
               df_edit.columns[9]:'Sangat_Sulit_CI95%_2',
               df_edit.columns[10]:'N'
                        }
              )
```
